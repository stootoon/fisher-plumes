```
reload(corrm)
Params     = corrm.IntermittentGeneralizedInverseGaussian.Params
TrueModel  = corrm.IntermittentGamma
true_model = TrueModel(name="True Model", init_params = Params(=2, =0.5,  = 0.01,  = 0.8, k = 0.4, m = 0.1,  = 0.1,  = 0.1),
                       _penalty = 0, _pr_mean = 0.8, _pr_strength = 1)
data, labs = true_model.gen_data(1001, params = true_model.params)
corrm.plot_data(data, labs)
```

```
reload(corrm)
Exp = corrm.CorrModel(model = "exp", name = "Exp", intermittent = True,
                      #init_params = Params(=2, =0.5,  = 0.01,  = 0.8, k = 0.2, m = 0.1,  = 0.1,  = 0.1),
                      _penalty = 0, _pr_mean = 0.8, _pr_strength = 1)
Gam = corrm.CorrModel(model = "gamma", name = "Gam", intermittent = True,
                      #init_params = Params(=2, =0.5,  = 0.01,  = 0.8, k = 0.5, m = 1.,  = 0.1,  = 0.1),
                      _penalty = 0, _pr_mean = 0.8, _pr_strength = 1)
Gig = corrm.CorrModel(model = "gig", name="Gig",intermittent = True,
                      #init_params = Params(=2, =0.5,  = 0.01,  = 0.8, k = 0.5, m = 0.2,  = 0.1,  = 0.1),
                      _penalty = 0, _pr_mean = 0.8, _pr_strength = 1, gammaness_thresh = 1e5)
```
