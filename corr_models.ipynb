```
def get_data(ds, idist, ifreq, iprb=0):
    F = data[ds]
    dd= array(list(F.pairs_um.keys()))
    dd= dd[dd>=0]
    d = dd[idist]
    f = ifreq
    print(f"Returning data for ds={ds}, d={d}, freq={F.freqs[f]:.1f}")
    y  = F.rho[iprb][d][0][f]
    la = F.la[iprb][d][0,f,0]
    mu = F.mu[iprb][d][0,f,0]
    print(f"la={la:.2g}, mu={mu:.2g}, {std(y)=:.2g}")
    la_ = la/std(y)
    mu_ = mu/std(y)
    y_  =  y/std(y)
    print(f"la_={la_:.2g}, mu_={mu_:.2g}, {std(y_)=:.2g}")
    return y_
```

# Model selection

```
from sklearn.model_selection import GridSearchCV, ShuffleSplit
import pandas as pd
```

```
param_grid= [{"model":["exp", "gamma", "gig"], "intermittent":[False]},
             {"model":["exp", "gamma", "gig"], "intermittent":[True], "γ_pr_mean":[0.2,0.4,0.6,0.8], "γ_pr_strength":[1]}]
```

```
reload(corrm)
corrm.logger.setLevel(logging.WARN)
```

```
search = GridSearchCV(estimator = corrm.CorrModel(name="Model", scoring = "tv", init_mode="ancestral"), param_grid=param_grid, cv=ShuffleSplit(n_splits=3, test_size=0.5, random_state=0),verbose=1, n_jobs=1) 
```

```
y_ = get_data("16Ts", 10, 10)
plot(y_)
```

```
search.fit(X=y_, y=None)
```

```
results_df = pd.DataFrame(search.cv_results_)
filtered_results = results_df[
    ['param_model', 'param_intermittent', 'param_γ_pr_mean', 'mean_test_score', 'std_test_score']
]
print(filtered_results.sort_values(by='mean_test_score', ascending=False).to_string(index=False))
```
