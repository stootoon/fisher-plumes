# CorrModel - Testing the wrapper class.

```
reload(corrm)
TrueModel  = corrm.IntermittentExponential
true_model = TrueModel(init_params = TrueModel.Params(=2, =0.5,  = 0.01,  = 0.8),
                       _penalty = 0, _pr_mean = 0.8, _pr_strength = 1)
data, labs = true_model.gen_data(1001, params = true_model.params)
corrm.plot_data(data, labs)
```

```
reload(corrm)
Params = corrm.IntermittentGeneralizedInverseGaussian.Params
Exp = corrm.CorrModel(model = "exp", intermittent = True,
    init_params = Params(=2, =0.5,  = 0.01,  = 0.8, k = 0.2, m = 0.1,  = 0.1,  = 0.1),
                    _penalty = 0, _pr_mean = 0.8, _pr_strength = 1)
Gam = corrm.CorrModel(model = "gamma", intermittent = True,
    init_params = Params(=2, =0.5,  = 0.01,  = 0.8, k = 0.5, m = 1.,  = 0.1,  = 0.1),
                    _penalty = 0, _pr_mean = 0.8, _pr_strength = 1)
Gig = corrm.CorrModel(model = "gig", intermittent = True,
    init_params = Params(=2, =0.5,  = 0.01,  = 0.8, k = 0.5, m = 0.2,  = 0.1,  = 0.1),
                      _penalty = 0, _pr_mean = 0.8, _pr_strength = 1)
```

```
Exp.fit(data);
Gam.fit(data, method = "TNC");
Gig.fit(data, method = "TNC");
```

```
corrm.plot_cdfs(data, [Exp, Gam, Gig])
```
