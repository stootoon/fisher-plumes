```
def get_data(ds, idist, ifreq, iprb=0):
    F = data[ds]
    dd= array(list(F.pairs_um.keys()))
    dd= dd[dd>=0]
    d = dd[idist]
    f = ifreq
    print(f"Returning data for ds={ds}, d={d}, freq={F.freqs[f]:.1f}")
    y  = F.rho[iprb][d][0][f]
    la = F.la[iprb][d][0,f,0]
    mu = F.mu[iprb][d][0,f,0]
    print(f"la={la:.2g}, mu={mu:.2g}, {std(y)=:.2g}")
    la_ = la/std(y)
    mu_ = mu/std(y)
    y_  =  y/std(y)
    print(f"la_={la_:.2g}, mu_={mu_:.2g}, {std(y_)=:.2g}")
    return y_
```

```
def get_length_scale(params):
    λ = params.λ if not hasattr(params, "α") else 2*params.λ/params.α
    μ = params.μ if not hasattr(params, "β") else 2*params.μ/params.β
    return [λ, μ]
```

```
def cleanup_cdf(c):
    # Replace any nan indices using linear interpolation
    c = np.array(c)
    nans = np.isnan(c)
    if nans.sum() == 0:
        return c
    else:
        print(f"Found {nans.sum()} nans.")
    x    = lambda z: z.nonzero()[0]
    c[nans] = np.interp(x(nans), x(~nans), c[~nans])
    return c

x = sin(2*pi*arange(100)/100)
x[50:80] = np.nan
plot(cleanup_cdf(x))
plot(x, linewidth=4)
```
