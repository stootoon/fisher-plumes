```
def get_data(ds, idist, ifreq, iprb=0):
    F = data[ds]
    dd= array(list(F.pairs_um.keys()))
    dd= dd[dd>=0]
    d = dd[idist]
    f = ifreq
    print(f"Returning data for ds={ds}, d={d}, freq={F.freqs[f]:.1f}")
    y  = F.rho[iprb][d][0][f]
    la = F.la[iprb][d][0,f,0]
    mu = F.mu[iprb][d][0,f,0]
    print(f"la={la:.2g}, mu={mu:.2g}, {std(y)=:.2g}")
    la_ = la/std(y)
    mu_ = mu/std(y)
    y_  =  y/std(y)
    print(f"la_={la_:.2g}, mu_={mu_:.2g}, {std(y_)=:.2g}")
    return y_
```

```
def find_best_(obj):
    w = Wrapper(obj)
    v, k = w.mean_test_score(model="exp")    
    vals = [1 if "_pr_mean" not in k[i] else k[i]["_pr_mean"] for i in range(len(k))]
    imax = argsort(-np.array(v))[0]
     = [m.models[-1].params. for m in w.models(**k[imax])[0][0]]
    return np.mean()
```

```
def find_best_non_intermittent(obj):
    w = Wrapper(obj)
    m, k = w.models(intermittent=False)
    s = np.array([w.mean_test_score(**ki)[0] for ki in k])
    ibest = argmax(s)
    m = m[ibest]
    k = k[ibest]
    s = s[ibest]
    return m,k,s
```

```
def find_k(obj):
    w = Wrapper(obj)
    m,k,s = find_best_non_intermittent(obj)
    return 1 if k["model"]=="exp" else np.nanmean([mi.models[-1].params.k for mi in m])

def find_gig_a(obj):
    w = Wrapper(obj)
    m,k,s = find_best_non_intermittent(obj)
    if k["model"] in ["exp", "gamma"]:
        return np.array([1/np.nanmean([mi.models[-1].params. for mi in m]), 1/np.nanmean([mi.models[-1].params. for mi in m])])
    else:
        return np.array(
            [np.nanmean([mi.models[-1].params./mi.models[-1].params. for mi in m]),
             np.nanmean([mi.models[-1].params./mi.models[-1].params. for mi in m])])

def find_gig_b(obj):
    w = Wrapper(obj)
    m,k,s = find_best_non_intermittent(obj)
    if k["model"] in ["exp", "gamma"]:
        return np.array([0, 0])
    else:
        return np.array([np.nanmean([mi.models[-1].params. * mi.models[-1].params. for mi in m]),
                np.nanmean([mi.models[-1].params. * mi.models[-1].params. for mi in m])])    
```
