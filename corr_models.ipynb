# Intermittent Generalized Inverse Gausssian model

```
reload(corrm)
Model = corrm.IntermittentGeneralizedInverseGaussian
```

```
true_params = Model.Params(=2,=1,=0.01,=0.8, k=0.2,m=0.1,=0.1, =0.1)
y, labs = Model.gen_data(1001, true_params)
corrm.plot_data(y, labs)
```

```
reload(corrm)
Model = corrm.IntermittentGeneralizedInverseGaussian
true_mdl = Model(true_params)
mdl = Model(Model.Params(=1,=0.1, k=1,m=1,=0.5, =0.5,  = 0.8,  = 0.01),
            _penalty = 0, _pr_mean = 0.8, _pr_strength = 1)
mdl.fit(y, method="Nelder-Mead", damping=0.1)
figure(figsize=(8,4))
corrm.plot_cdfs(y, [true_mdl, mdl])
```
